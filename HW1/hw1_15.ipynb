```python
import numpy as np
import random as rd

def sign(xt,wt):
    if np.dot(xt,wt)>0:
        return 1
    else:
        return -1

def InputData(fName):
    f=open(fName)
    text=f.readlines()
    X=[]
    Y=[]
    c=0
    for r in text:
        row=r.strip()
        row=row.replace('\t',' ')
        l=row.split(' ')
        X.append([])
        for i in range(4):
    #         print(l[i])
            X[c].append(float(l[i]))
        Y.append(int(l[4]))
        c+=1
    f.close()
    return X,Y,c
    
def PLA_Naive(X,Y,W,iteration):
    flag=False
    updates=0
    for i in range(iteration):
        flag=False
        for lp in range(len(X)):
            if sign(X[lp],W) != Y[lp]:
                W=W+Y[lp]*X[lp]
                flag=True
                updates+=1
        if not flag:
            print("WIN!!",i)
            break
    return W,updates

def PLA_Random(X,Y,W,iteration):
    flag=False
    updates=0
    v_n=0.5
    Id=[]
    for i in range(len(Y)):
        Id.append(i)
    
    Idt=Id
    rd.shuffle(Idt)
        
    for j in range(iteration):
        flag=False
        for lp in Idt:
            if sign(X[lp],W) != Y[lp]:
                W=W+Y[lp]*X[lp]
#                 a=Y[lp]*X[lp]
#                 a[1::]=a[1::]*v_n
#                 W=W+a
                flag=True
                updates+=1
        if not flag:
#             print("WIN!!",j)
            n_inter=j
            return updates,j
    return updates,999999    

def PLA_ErrorCount(X,Y,W_t):
    Error=0
    for i in range(len(Y)):
        if sign(X[i],W_t) != Y[i]:
            Error+=1
    return Error

def PLA_Pocket(X,Y,W,iteration):
    flag=False
    updates=0
    W_best=W
    W50=W
    Id=[]
    for i in range(len(Y)):
        Id.append(i)
    
    Idt=Id
    rd.shuffle(Idt)
    
    Error=PLA_ErrorCount(X,Y,W)
    for j in range(iteration):
        flag=False
        for lp in Idt:
            if sign(X[lp],W) != Y[lp]:
                W=W+Y[lp]*X[lp]
                Error_t=PLA_ErrorCount(X,Y,W)
                flag=True
                updates+=1
                if Error_t<Error:
                    Error=Error_t
                    W_best=W
        if not flag:
            print("WIN!!",j)
            n_inter=j
            break
        if updates>=100:
#             print("Update is up to 50 times!")
#             W50=W
            break
    return W_best,updates   

# def PLA_Verify(X_test,Y_test,W):
    

X=[]
Y=[]
X_test=[]
y_test=[]
X,Y,TrainC=InputData('hw1_18_train.dat')
X_test,Y_test,TestC=InputData('hw1_18_test.dat')
X=np.array(X)
Y=np.array(Y)
X=np.concatenate((np.ones((len(X),1)),X),axis=1)
print(TrainC)
X_test=np.array(X_test)
Y_test=np.array(Y_test)
X_test=np.concatenate((np.ones((len(X_test),1)),X_test),axis=1)
W=np.zeros(5) #shift the W[0] makes little influence on updating times

# for i in range(5):
#     print(X[i],Y[i])
# W,updates=PLA_Naive(X,Y,W,200000)
# print(updates)

# Sum_update=0
# for i in range(2000):
#     updates,loops=PLA_Random(X,Y,W,20000)
#     Sum_update+=updates
# print(Sum_update/2000,Sum_update)
# 

# Sum_err=0
# for i in range(2000):
# #     print(i)
#     W_best,updates=PLA_Pocket(X,Y,W,50)
#     Err=PLA_ErrorCount(X_test,Y_test,W_best)/TestC
# #     print(Err)
#     Sum_err+=Err
# print(Sum_err/2000)

Sum_err=0
for i in range(2000):
#     print(i)
    W_best,updates=PLA_Pocket(X,Y,W,100)
    Err=PLA_ErrorCount(X_test,Y_test,W_best)/TestC
#     print(Err)
    Sum_err+=Err
print(Sum_err/2000)
```
